```python
import json
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .master("spark://192.168.5.177:7077") \
    .appName("Group36-Reddit-Final-Project") \
    .config("spark.driver.host", "192.168.5.177") \
    .config("spark.driver.bindAddress", "0.0.0.0") \
    .config("spark.dynamicAllocation.enabled", "true") \
    .config("spark.shuffle.service.enabled", "false") \
    .config("spark.dynamicAllocation.shuffleTracking.enabled", "true") \
    .getOrCreate()

sc = spark.sparkContext

reddit_rdd = sc.textFile("hdfs://192.168.5.177:9000/user/ubuntu/reddit_data/corpus-webis-tldr-17.json")

first_line = reddit_rdd.first()
sample_json = json.loads(first_line)

print("--- Detected Schema Keys ---")
for key in sample_json.keys():
    print(f"Key: {key} | Type: {type(sample_json[key]).__name__}")

total_records = reddit_rdd.count()

print(f"\n--- Dataset Volume ---")
print(f"Total Number of Records: {total_records:,}")
```
